```python
password   = 483503

empty_data = 0.0

waktu      = 24
hari       = 30

dingin     = 0
sedang     = 0
panas      = 0

def input_password() :
  return int(input('Masukkan password anda : '))

def konversi_suhu(suhu,satuan) :
  while(1) :
    if(satuan == '3') : 
      return suhu - 273.15
      break
    elif(satuan == '2') :
      return (suhu - 32)*(5/9)
      break
    elif(satuan == '1') :
      return suhu
      break
    else :
      print('Silahkan masukkan angka 1/2/3')

#### password ####

while(1) :
    
  if(input_password() != password) :
    print('Password keliru, silahkan masukkan password anda kembali')
  else :
    print('password benar, silahkan lanjutkan mengisi data')
    break

#### membuat database ####

database = [[empty_data for i in range(waktu)] for j in range(hari)]

print(database)

#### mengisi database ####

while (1) :

  tanggal = int(input('Masukkan tanggal : '))
  jam = int(input('Masukkan jam : '))
  suhu = float(input('Masukkan suhu : '))
  satuan = str(input('Masukkan satuan 1.Celcius 2.Farenheit 3. Kelvin : '))
  hasil_konversi = konversi_suhu(suhu,satuan)
  hasil_konversi = "{:.2f}".format(hasil_konversi)

  print('\ndata yang dimasukkan \ntanggal \t:',tanggal,'\njam \t\t:',jam,'\nsuhu \t\t:',suhu,'\nsatuan \t\t:',satuan,'\nhasil konversi \t:',hasil_konversi,'\n')

  database[tanggal - 1][waktu - 1] = hasil_konversi

  flag = str(input('apakah ada data lain yang akan dimasukkan?(y/n)'))

  if flag == 'n' : break

print(database)

#### counter ####

for j in range (waktu) :
  for k in range(hari) :

    data_suhu = float(database[k][j])

    if data_suhu != 0.0 and data_suhu < 20 :
      dingin = dingin + 1
    elif data_suhu != 0.0 and data_suhu >= 25 :
      panas = panas + 1
    elif data_suhu != 0.0 and (data_suhu >= 20 or data_suhu < 25) :
      sedang = sedang + 1

print('Kondisi hari panas : ',panas)
print('Kondisi hari sedang : ',sedang)
print('Kondisi hari dingin : ',dingin)
```
