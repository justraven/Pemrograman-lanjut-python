```python
tanggal = int(input('Masukkan tanggal : '))
jam = int(input('Masukkan jam : '))
Suhu = float(input('Masukkan suhu : '))
Satuan = str(input('Masukkkan satuan suhu : 1.Celcius / 2. Farenheit / 3. Kelvin'))


print('Tanggal : ',tanggal)
print('Jam : ',jam)
print('Suhu : ',Suhu)
print('Satuan suhu : ',Satuan)

while(1) :
    if(Satuan == '3') :
        Suhu = Suhu - 273.15 #konversi kelvin ke celcius
        break
    elif(Satuan == '2') :
        Suhu = (Suhu - 32)*(5/9) #konversi farenheit ke celcius
        break
    elif (Satuan == '1') :
        Suhu = Suhu
        break
    else :
        print('Silahkan masukkan angka 1/2/3')

Suhu = "{:.2f}".format(Suhu)
print('Hasil konversi suhu : ',Suhu)
```
